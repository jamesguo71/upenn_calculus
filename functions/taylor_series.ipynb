Are other functions similar to $e$, $cos$, and $sin$, in terms of being able to be expanded into a serie?

We may want to assume "every" "reasonable" function can be expressed in some form of sum of series

$
\begin{equation}
\begin{split}
f(x) &= \sum_{k=0}^{\infty} C_k x^k
&= C_0 + C_1 x + C_2 x^2 + C_3 x^3 + ...
\end{split}
\end{equation}
$

... for some $\{C_k\}$

This is not true. We need to be careful about what we mean by "every" and "reasonable".

But for now, let's pretend it's true, and think of how to compute $\{C_k\}$.

# Definition of Taylor Series

The definition of the Taylor Series of $f(x)$ at $x = 0$ is

$
\begin{equation}
\sum_{k=0}^{\infty} \frac{f^{(k)}(0)}{k!} x^k = f(0) + \frac{df}{dx}|_0 x + \frac{1}{2!} \frac{d^2f}{dx^2}|_0 x^2 + \frac{1}{3!} \frac{d^3f}{dx^3}|_0 x^3 + ... \\
\end{equation}
$

That is,
$
C_k = \frac{f^{(k)}(0)}{k!} = \frac{1}{k!} \frac{d^kf}{dx^k}|_0
$

## Examples

Let $f(x) = e^x$

We have $\begin{equation} \frac{d^ne}{dx^n} = e \end{equation}$
and $\begin{equation} e^0 = 1 \end{equation}$

$
\begin{equation}
\begin{split}
e^x &= 1 + x + \frac{1}{2!} x^2 + \frac{1}{3!} x^3 + \frac{1}{4!} x^4 + \frac{1}{5!} x^5 + ...\\
&= e^0 + \frac{de}{dx}|_0 x + \frac{1}{2!} \frac{d^2e}{dx^2}|_0 x^2 + \frac{1}{3!} \frac{d^3e}{dx^3}|_0 x^3 + \frac{1}{4!} \frac{d^4e}{dx^4}|_0 x^4 + \frac{1}{5!} \frac{d^5e}{dx^5}|_0 x^5 + ...\\
\end{split}
\end{equation}
$

Let $f(x) = \sin x$


We have

$
\begin{array}{l}
\frac{d}{dx} \sin &=& \cos \\
\frac{d}{dx} \cos &=& -\sin \\
\frac{d}{dx} - \sin &=& -\cos \\
\frac{d}{dx} - \cos &=& \sin \\
\end{array}
$

...which is cyclic with a frequency of 4.

and $\begin{equation} \sin 0 = 0 \end{equation}$

$
\begin{equation}
\begin{split}
\sin x &= x - \frac{1}{3!} x^3 + \frac{1}{5!} x^5 - ... \\
&= 0 + 1 x + 0 \frac{1}{2!} x^2 + (-1) \frac{1}{3!} x^3 + 0 \frac{1}{4!} x^4 + 1 \frac{1}{5!} x^5 + ... \\
&= \sin 0 + (\cos 0) x  + (-\sin 0) \frac{1}{2!} x^2 + (-\cos 0) \frac{1}{3!} x^3 + (\sin 0) \frac{1}{4!} x^4  + (\cos 0) \frac{1}{5!} x^5 + ...\\
&= \sin 0 + \frac{d\sin}{dx}|_0 x + \frac{1}{2!} \frac{d^2\sin}{dx^2}|_0 x^2 + \frac{1}{3!} \frac{d^3\sin}{dx^3}|_0 x^3 + \frac{1}{4!} \frac{d^4\sin}{dx^4}|_0 x^4 + \frac{1}{5!} \frac{d^5\sin}{dx^5}|_0 x^5 + ...
\end{split}
\end{equation}
$

Let $f(x) = \cos x$

We have

$
\begin{array}{l}
\frac{d}{dx} \cos &=& -\sin \\
\frac{d}{dx} -\sin &=& -\cos \\
\frac{d}{dx} - \cos &=& \sin \\
\frac{d}{dx} \sin &=& \cos \\
\end{array}
$
...which is cyclic with a frequency of 4.

and $\begin{equation} \cos 0 = 1 \end{equation}$

$
\begin{equation}
\begin{split}
\cos x &= 1 - \frac{1}{2!} x^2 + \frac{1}{4!} x^4 - ... \\
&= 1 + 0 x + (-1) \frac{1}{2!} x^2 + 0 \frac{1}{3!} x^3 + 1 \frac{1}{4!} x^4 + ... \\
&= \cos 0 + (-\sin 0) x  + (-\cos 0) \frac{1}{2!} x^2 + (\sin 0) \frac{1}{3!} x^3 + (\cos 0) \frac{1}{4!} x^4  + ...\\
&= \cos 0 + \frac{d\cos}{dx}|_0 x + \frac{1}{2!} \frac{d^2\cos}{dx^2}|_0 x^2 + \frac{1}{3!} \frac{d^3\cos}{dx^3}|_0 x^3 + \frac{1}{3!} \frac{d^4\cos}{dx^4}|_0 x^4 + ...
\end{split}
\end{equation} \frac{d}{dx} \cos 0
$

Let $f(x) = x^3 + 2x^2 - x + 5$

$
\begin{equation}
\begin{split}
x^3 + 2x^2 - x + 5 &= f(0) + \frac{df}{dx}|_0 x + \frac{1}{2!} \frac{d^2f}{dx^2}|_0 x^2 + \frac{1}{3!} \frac{d^3f}{dx^3}|_0 x^3 + ... \\
&= 5 + (3(0^2) + 4(0) - 1) x + \frac{1}{2!} (6(0) + 4) x^2 + \frac{1}{3!} (6) x^3 + \frac{1}{3!} 0 x^4 + ... \\
&= 5 - x + 2 x^2 + x^3 \\
&= x^3 + 2x^2 - x + 5
\end{split}
\end{equation}
$

Polynomials have themselves as their Taylor Series.

## Hard functions -> Polynomials
Taylor Expansion helps us to convert obtuse, difficult, and unknown functions into simple polynomials. 

### Some functions are best expressed as Taylor Series
#### Bessel Function

```python
from sympy import symbols, Eq, Sum, oo, init_printing, Rational, factorial, Pow

init_printing()

x, k, J0, n = symbols('x k J_{0}, n')

bessel_j0_term = Pow(-1, k) * (Pow(x ,2 * k) / (Pow(2, (2 * k)) * Pow(factorial(k), 2)))

Eq(
    J0,
    Sum(
        bessel_j0_term,
        (k, oo, n),
    )
)
```

```python
%matplotlib inline
from sympy import plot

def approximate_bessel_j0(num_of_terms):
    return sum(bessel_j0_term.subs(k, i) for i in range(num_of_terms))

approximate_bessel_j0(10)

for i in range(15):
    plot(approximate_bessel_j0(i), xlimit=(-5, 5), ylimit=(-5, 5))
```

### Modeling rotating rope with Bessel function

```python
from sympy import sqrt

r = symbols('r')  # Displacement of a point on the rope
G = symbols('G')  # Gravitational constant
omega = symbols('omega')  # Angular frequency
x = symbols('x')  # Distance from bottom
Eq(r, J0((2 * omega / sqrt(G)) * sqrt(x)) )
```
