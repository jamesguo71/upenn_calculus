What happen when a function gets very, very small, or very, very large?

How fast does a function go to $0$? How quickly to $\infty$?

Quantify these rates of change will give us a new language, the big $O$, or asymptotic notation.

## Examples of functions growing to $\infty$, but in different rates



#### $e^x$ vs $x^n$

$$\lim_{x \to \infty} x^n = \infty$$ and $$\lim_{x \to \infty} e^x$$.

Both functions grow to $\infty$ as $x$ approaches to $\infty$. But how do they compare in terms of the growth rate?

We can compare the rate of growth of the two functions by computing the limit of the quotient

$$
\begin{equation}
\begin{aligned}{}
\lim_{x \to \infty} \frac{x^n}{e^x} &= \lim_{x \to \infty} \frac{n x^{n - 1}}{e^x} & \text{applying L'Hopital's Rule} \\
. \\
. \\
. & \text{(keep applying L'Hopital's Rule)} \\
&= \lim_{x \to \infty} \frac{n!}{e^x} \\
&= 0
\end{aligned}
\end{equation}
$$

This means $e^x$ grows faster than $x^n$, hence the exponential function grows faster than any monomial (and also polynomial) function.

####  $x$ vs $(\ln x)^n$

Similarly, we can compare the rate of growth of $(\ln x)^n$ vs $x$ by computing the limit of the quotient of the two functions as x approaches $\infty$.

$$
\begin{equation}
\begin{aligned}{}
\lim_{x \to \infty} \frac{(\ln x)^n}{x} &= \lim_{x \to \infty} \frac{n (\ln x)^{n - 1}}{x} & \text{applying L'Hopital's Rule} \\
. \\
. \\
. & \text{(keep applying L'Hopital's Rule)} \\
&= \lim_{x \to \infty} \frac{n!}{x} \\
&= 0
\end{aligned}
\end{equation}
$$

This means $x$ grows faster than $(\ln x)^n$, hence monomial (polynomial) grow is faster than (poly) logarithmic grow.

Hence exponential beats polynomial beats logarithmic.

### Hierarchy of rate of growth of functions as $x \to \infty$

- Factorial
- Exponential
- Polynomial
- Logarithmic
- Constant

### Visualizing growth of $x^n$ for different $n$s ($n > 0$). 

```python
from sympy import init_printing, symbols, pi, sin, cos, atan, ln, sqrt, exp, sinh, cosh, factorial
from sympy.plotting import plot

init_printing()

colors = ['black', 'blue', 'brown', 'green', 'violet', 'red']

plot_kwargs = {
    'xlim': (0, 10),
    'ylim': (0, 10),
    'legend': True,
}

x = symbols('x')

monomial_plots = [plot(x ** n, line_color=colors.pop(), **plot_kwargs) for n in range(1, 7)]

zoom_in_plot_kwargs = {
    'xlim': (0, 1),
    'ylim': (0, 1),
    'legend': True,
}

sin_plot = plot(sin(x), **zoom_in_plot_kwargs)
ln_plot = plot(ln(1 + x), **zoom_in_plot_kwargs)
cos_plot = plot(1 - cos(x ** 2), **zoom_in_plot_kwargs)
sqrt_plot = plot(sqrt(x), **zoom_in_plot_kwargs)
exp_plot = plot(exp(-1 / (x ** 2)), **zoom_in_plot_kwargs)

pi_by_two_constant_plot = plot(pi/2, line_color='aqua', **plot_kwargs)
atan_plot = plot(atan(x), **plot_kwargs)
x_three_by_two_plot = plot(x ** (3 / 2), line_color='aqua', **plot_kwargs)
x_sqrt_plot = plot(x * sqrt(1 + x ** 2), **plot_kwargs)
compare_ln_plot = plot(ln(x), line_color='aqua', **zoom_in_plot_kwargs)
ln_sinh_plot = plot(ln(sinh(x)), **plot_kwargs)
compare_exp_plot = plot(exp(x), line_color='aqua', **plot_kwargs)
cosh_plot = plot(cosh(x), **plot_kwargs)
compare_5_dot_1_ln_plot = plot(5.1 * ln(x), line_color='aqua', **zoom_in_plot_kwargs)
ln_x_5_plot = plot(ln(x ** 5), **plot_kwargs)
x_to_x_plot = plot(x ** x, **plot_kwargs)

aggregated_monomial_plot = monomial_plots.pop()
for plot_ in monomial_plots:
    aggregated_monomial_plot.extend(plot_)
```

```python
%matplotlib inline

aggregated_monomial_plot.show()
```

As $x \to \infty$, the larger $n$ is, the faster the function __grows__.

Hence $x^n < x ^{n + 1}$.

```python
%matplotlib inline

## Zoom-in the interval at [0, 1]
aggregated_monomial_plot.xlim = (0, 1)
aggregated_monomial_plot.ylim = (0, 1)
aggregated_monomial_plot.show()
```

As $x \to 0$, the larger $n$ is, the faster the function __shrinks__.

Hence $x^n > x ^{n + 1}$.

#### Examples

$$
\begin{equation}
\begin{aligned}{}
\lim_{x \to 0} \frac{1 - 2 x + 3 x^2}{4 - 5 x + 6 x^2} &= \frac{\lim_{x \to 0}(1 - 2 x + 3 x^2)}{\lim_{x \to 0}(4 - 5 x + 6 x^2)} \\
&= \frac{1}{4}
\end{aligned}
\end{equation}
$$

Looking at the limit of the same function but for $x \to \infty$

$$
\begin{equation}
\begin{aligned}{}
\lim_{x \to \infty} \frac{1 - 2 x + 3 x^2}{4 - 5 x + 6 x^2} &=\lim_{x \to \infty} \frac{\frac{1}{x^2} - \frac{2}{x} + 3}{\frac{4}{x^2} - \frac{5}{x} + 6} \\
&= \frac{3}{6} \\
&= \frac{1}{2}
\end{aligned}
\end{equation}
$$

As $x \to 0$ it is the lowest order terms that dominate; where as $x \to \infty$ it is the highest order terms that dominate.

This duality is extremely important and can cause confusion. Always pay attention to which way the limit is going.

## Big $O$: the language for controlling growth

### Definition

$f$ is in $O(x^n)$ if $|f(x)| < C|x^n|$ for some $C$.

This definition just gives an upper bound (how bad it can be). It does not actually care about the value. It is just for controlling the growth.

Two versions of this definition:

1. This inequality has to hold as $x \to 0$. E.g. $O(x^n)$ consists all functions that approach $0$ at least as fast as $x^n$.
2. This inequality has to hold as $x \to \infty$. E.g. $O(x^n)$ consist all functions that approach $\infty$ no faster than $x^n$.

### H.O.T. (Higher Order Terms) is now replaced by the big $O$

For example, the Taylor's series of $\arctan x$ about $x = 0$ will be

$
\begin{equation}
\begin{aligned}{}
\arctan x &= x - \frac{1}{3} x^3 + \frac{1}{5} x^5 - ... \\
&= x - O(x^2) \\
&= x - O(x^3) & \text{to be more precise} \\
&= x - \frac{1}{3} x^3 + O(x^5) & \text{even further}
\end{aligned}
\end{equation}
$

Likewise, as $x \to \infty$

$
\begin{equation}
\begin{aligned}{}
x \sqrt{5 + 3 x + x^2} &= x^2 + O(x) \\
&= x^2 + \frac{3}{2} x + O(1) & \text{more specific from the binomial series}
\end{aligned}
\end{equation}
$

We can generalize the big $O$ definition

### A more general definition of big $O$

$f$ is in $O(g(x))$ if $|f(x)| < C|g(x)|$ for some $C$.

### More examples

#### As $x \to 0$

- $5 x + 3 x^2$ is in $O(x)$ but is not in $O(x^2)$
- $\sin x$ is in $O(x)$ but is not in $O(x^2)$
- $\ln(1 + x) - x$ is in $O(x^4)$ but is not in $O(x^5)$
- $\sqrt{x}$ is not in $O(x^n)$ for any $n \geq 1$
- $e^{-\frac{1}{x^2}}$ is in $O(x^n)$ for all $n \geq 1$

## Asymptotic analysis

### Rules

- $O(x^m) \cdot O(x^n) = O(x^{m + n})$ as $x \to 0$ or $x \to \infty$.
- $
\begin{equation}
O(x^m) + O(x^n) = \begin{cases}
O(x^{\min \{m, n\}}) & \text{as $x \to 0$} \\
O(x^{\max \{m, n\}}) & \text{as $x \to \infty$} \\\end{cases}
\end{equation}
$

### Example

#### As $x \to 0$

$\begin{equation}
\begin{aligned}{}
\frac{e^{x}}{1 - x} &= e^{x} \frac{1}{1 - x} \\
&= (1 + x + O(x^2))(1 + x + O(x^2)) \\
&= ((1 + x) + O(x^2))((1 + x) + O(x^2)) \\
&= (1 + x)^2 + 2 (1 + x) O(x^2) + (O(x^2))^2 \\
&= (1 + 2 x + x^2) + 2 (1 + x) O(x^2) + (O(x^2))^2 \\
&= (1 + 2 x + O(x^2)) + 2 (1 + x) O(x^2) + (O(x^2))^2 \\
&= 1 + 2 x + O(x^2) + 2 O(x^2) + 2 x O(x^2) + O(x^4)\\
&= 1 + 2 x + O(x^2) + O(x^2) + O(x^3) + O(x^4) \\
&= 1 + 2 x + (O(x^2) + O(x^2) + O(x^3) + O(x^4)) \\
&= 1 + 2 x + O(x^2) & \text{as $O(x^m) + O(x^n) = O(\min\{m, n\})$ as $x \to 0$} \\
\end{aligned}
\end{equation}
$

We can make it more interesting

$\begin{equation}
\begin{aligned}{}
(\frac{e^{x}}{1 - x})^\alpha &= (1 + 2 x + O(x^2))^\alpha \\
&= 1 + 2 \alpha x + O(x^2) & \text{binomial series expansion}
\end{aligned}
\end{equation}
$
