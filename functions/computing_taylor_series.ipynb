Recall the definition of taylor series of $f(x)$ about $x=0$

$
\begin{equation}
f(x) = \sum_{k=0}^{\infty} \frac{f^{(k)}(0)}{k!} x^k = f(0) + \frac{df}{dx}|_0 x + \frac{1}{2!} \frac{d^2f}{dx^2}|_0 x^2 + \frac{1}{3!} \frac{d^3f}{dx^3}|_0 x^3 + ... \\
\end{equation}
$

...which involves a lot of derivatives.

That's not your only choice. There are few other techniques like substition and combine like terms. 

## Substition method

Consider a function

$
\frac{1}{x} \sin x^2
$

...which differentiating this function will be a lot of work.

However, we do know what $\sin x$ is in terms of Taylor series:

$
\begin{equation}
\sin x = x - \frac{1}{3!} x^3 + \frac{1}{5!} x^5 - ...
\end{equation}
$

By substitute $x$ by $x^2$, we have

$
\begin{equation}
\begin{split}
\frac{1}{x} \sin x^2 &= \frac{1}{x} ((x^2) - \frac{1}{3!} (x^2)^3 + \frac{1}{5!} (x^2)^5 - ...) \\
&= x - \frac{1}{3!} x^5+ \frac{1}{5!} x^9 - ... \\
&= x^{-1} \sum_{k=0}^{\infty} (-1)^k \frac{(x^2)^{2k+1}}{(2k+1)!} \\
&= \sum_{k=0}^{\infty} (-1)^k \frac{x^{4k+1}}{(2k+1)!}
\end{split}
\end{equation}
$


## Combine like terms method

Consider a fucntion

$\cos^2 x$

We know that

$
\begin{equation}
\cos x = 1 - \frac{1}{2!} x^2 + \frac{1}{4!} x^4 - ...
\end{equation}
$

Hence

$
\begin{equation}
\begin{split}
\cos^2 x &= (\cos x) (\cos x) \\
&= (1 - \frac{1}{2!} x^2 + \frac{1}{4!} x^4 - ...) (1 - \frac{1}{2!} x^2 + \frac{1}{4!} x^4 - ...) \\
&= 1 - 2\frac{1}{2!}x^2 + (-\frac{1}{2!} x^2)^2 + 2(\frac{1}{4!}x^4) - \frac{1}{2!} x^2 \frac{1}{4!} x^4 - 2 \frac{1}{6!}x^6 + ... \\
&= 1 - x^2 + \frac{1}{3!} x^4 - \frac{2}{45} x^6 + ... \\
&= 1 - \sin^2 x
\end{split}
\end{equation}
$

Hence

$
\begin{equation}
\sin^2 x = x^2 - \frac{1}{3!} x^4 + \frac{2}{45} x^6 - ...
\end{equation}
$


## Hyperbolic Trigs

$
\begin{equation}
\cosh x = \frac{e^x + e^{-x}}{2}
\end{equation}
$

$
\begin{equation}
\sinh x = \frac{e^x - e^{-x}}{2}
\end{equation}
$

$
\begin{equation}
\begin{split}
\tanh x &= \frac{\sinh x}{\cosh h} \\
&= \frac{e^x - e^{-x}}{e^x + e^{-x}}
\end{split}
\end{equation}
$

### How do these functions look like?

```python
%matplotlib inline


from sympy import symbols, exp, init_printing, plot
from sympy.functions.elementary.trigonometric import cosh, sinh

init_printing()

x = symbols('x')

curve_cosh = plot(cosh(x))
curve_sinh = plot(sinh(x))

curve_cosh.append(curve_sinh[0])
curve_cosh.show()
```

### Hyperbolic Trigs Identities

$
\begin{equation}
\begin{split}
\cosh^2 x - \sinh^2 x &= (\frac{e^x + e^{-x}}{2})^2 - (\frac{e^x - e^{-x}}{2})^2 \\
&= \frac{e^{2x} + e^{-2x} + 2 \cdot e^x e^{-x}}{4} - \frac{e^{2x} + e^{-2x} - 2 \cdot e^x e^{-x}}{4} \\
&= \frac{4 \cdot e^x e^{-x}}{4} \\
&= e^x e^{-x} \\
&= 1
\end{split}
\end{equation}
$

#### Taylor series of hyperbolic functions

$
\begin{equation}
\begin{split}
\cosh x &= \frac{e^x + e^{-x}}{2} \\
&= \frac{1}{2} e^x + \frac{1}{2} e^{-x} \\
&= \frac{1}{2} (1 + x + \frac{1}{2!} x^2 + \frac{1}{3!} x^3 + \frac{1}{4!} x^4 + ...) + \frac{1}{2} (1 - x + \frac{1}{2!} x^2 - \frac{1}{3!} x^3 + \frac{1}{4!} x^4 - ...) \\
&= 1 + \frac{1}{2!} x^2 + \frac{1}{4!} x^4 + ... \\
&= \sum_{k=0}^{\infty} \frac{x^{2k}}{(2k)!}
\end{split}
\end{equation}
$

...by combining like terms

Just like the Taylor series of $\cos$, the Taylor series of $\cosh$ just contains even terms, but without alternating $(-1)$s.

This is a connection between trigometric functions and hyper trigometric functions.

Similarly

$
\begin{equation}
\begin{split}
\sinh x &= \frac{e^x - e^{-x}}{2} \\
&= \frac{1}{2} e^x - \frac{1}{2} e^{-x} \\
&= \frac{1}{2} (1 + x + \frac{1}{2!} x^2 + \frac{1}{3!} x^3 + \frac{1}{4!} x^4 - \frac{1}{5!} x^5 + ...) - \frac{1}{2} (1 - x + \frac{1}{2!} x^2 - \frac{1}{3!} x^3 + \frac{1}{4!} x^4 - \frac{1}{5!} x^5 ...) \\
&= x + \frac{1}{3!} x^3 + \frac{1}{5!} x^5 + ... \\
&= \sum_{k=0}^{\infty} \frac{x^{2k + 1}}{(2k + 1)!}
\end{split}
\end{equation}
$

... by combining like terms.

Just like the Taylor series of $\sin$, the Taylor series of $\sinh$ contains just odd terms, but without the alternating $(-1)$ s.

## Manipulating Taylor series as easy (or hard) as manipulating polynomial terms

$
\begin{equation}
\begin{split}
\frac{d}{dx} \sinh x &= \frac{d}{dx} \sum_{k=0}^{\infty} \frac{x^{2k + 1}}{(2k + 1)!} \\
&= \sum_{k=0}^{\infty} \frac{d}{dx} \frac{x^{2k + 1}}{(2k + 1)!} \\
&= \sum_{k=0}^{\infty} (2k + 1) \frac{x^{2k}}{(2k + 1)!} \\
&= \sum_{k=0}^{\infty} \frac{x^{2k}}{(2k)!} \\
&= \cosh x
\end{split}
\end{equation}
$

Similarly

$
\begin{equation}
\begin{split}
\frac{d}{dx} \cosh x &= \frac{d}{dx} \sum_{k=0}^{\infty} \frac{x^{2k}}{(2k)!} \\
&= \sum_{k=0}^{\infty} \frac{d}{dx} \frac{x^{2k}}{(2k)!} \\
&= \sum_{k=1}^{\infty} 2k \frac{x^{2k - 1}}{(2k)!} \\
&= \sum_{k=1}^{\infty} \frac{x^{2k - 1}}{(2k - 1)!} \\
&= \sum_{k=0}^{\infty} \frac{x^{2k + 1}}{(2k + 1)!} \\
&= \sinh x
\end{split}
\end{equation}
$

## H.O.T: Higher Order Terms

$
\begin{equation}
e^x = 1 + x + \frac{1}{2!} x^2 + \frac{1}{3!} x^3 + \frac{1}{4!} x^4 + \frac{1}{5!} x^5 + H.O.T
\end{equation}
$

$
\begin{equation}
\cos x = 1 - \frac{1}{2!} x^2 + H.O.T
\end{equation}
$

$
\begin{equation}
\begin{split}
\sin x &= x + H.O.T. \\
&= x - \frac{1}{3!} x^3 + H.O.T. \\
&= x - \frac{1}{3!} x^3 + - \frac{1}{5!} x^5 + H.O.T. \\
&= x - \frac{1}{3!} x^3 + - \frac{1}{5!} x^5 + \frac{1}{7!} x^7 + H.O.T.
\end{split}
\end{equation}
$

...and we can stop at whatever term we like.

### Example of the use of H.O.T.

Let's compute the Taylor series of

$
\begin{equation}
f(x) = 1 - 2x e^{\sin{x^2}}
\end{equation}
$

First compute $\sin x^2$

$
\begin{equation}
\begin{split}
\sin x &= x^2 - \frac{1}{3!} (x^2)^3 + H.O.T \\
&= x^2 - \frac{1}{3!} x^6 + H.O.T
\end{split}
\end{equation}
$

Then we have

$
\begin{equation}
\begin{split}
1 - 2x e^{\sin{x^2}} &= 1 + (x^2 - \frac{1}{3!} x^6 + H.O.T) + \frac{1}{2!}(x^2 + H.O.T)^2 + \frac{1}{3!} (x^2 + H.O.T)^3 + H.O.T \\
&= 1 + (x^2 - \frac{1}{6} x^6 + H.O.T) + \frac{1}{2}(x^2 + H.O.T)^2 + \frac{1}{6} (x^2 + H.O.T)^3 + H.O.T \\
&= 1 + x^2 + \frac{1}{2} x^4 + H.O.T
\end{split}
\end{equation}
$

Hence

$
\begin{equation}
\begin{split}
f(x) &= 1 - 2x e^{\sin{x^2}} \\
&= 1 - 2x (1 + x^2 + \frac{1}{2} x^4 + H.O.T) \\
&= 1 - 2x - 2x^3 - x^5 + H.O.T.
\end{split}
\end{equation}
$

Getting more terms will require more work.
