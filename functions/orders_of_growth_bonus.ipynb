We have discussed the hierarchy of orders of growth as $x \to \infty$

- Factorial
- Exponential
- Polynomial
- Logarithmic
- Constant

But what do we mean by factorial functions?

## Definition: Factorial ($\Gamma$ Function)

$$
\begin{equation}
x! = \int_{t = 0}^{\infty} t^{x} e^{-t} d t
\end{equation}
$$

This definition satisfies $x! = x (x - 1)!$.

## Factorial vs Exponential

$$
\begin{equation}
\begin{aligned}{}
\lim_{x \to \infty} \frac{x!}{e^x} &= \lim_{x \to \infty} \frac{x(x - 1)(x - 2) ... 3 \cdot 2 \cdot 1}{e \cdot e \cdot e ... \cdot e \cdot e \cdot e} \\
&= \lim_{x \to \infty} \frac{x}{e} \frac{x - 1}{e} \frac{x - 2}{e} ... \frac{3}{e} \frac{2}{e} \frac{1}{e}
&= \infty
\end{aligned}
\end{equation}
$$

## Big $O$: How is this language used? Why is this language used?

### Big $O$ is an efficient language:  Examples

#### Bounded function
$f(x)$ is in $O(1)$ $\Leftrightarrow$ $f$ is a bounded function.

#### Computation complexity
E.g. How much effort will it be to multiply two $n$ digits number? 

$$n \cdot 2 n + 2 n (n - 1) = O(n^2)$$.

#### Error analysis

We know the kinetic energy of a body is $E_k = \frac{1}{2} m v^2$, where $m$ is the mass and $v$ is the velocity;

But say if the measurement of velocity has some error ($\epsilon$), and is assumed very small ($\epsilon \to 0$). 

What will be the induced error in the kinetic energy? 

$$
\begin{equation}
\begin{aligned}{}
E_k &= \frac{1}{2} m (v + \epsilon)^2 \\
&= \frac{1}{2} m v^2 + m v \epsilon + \frac{1}{2} m \epsilon^2 \\
&= \frac{1}{2} m v^2 + m v \epsilon + O(\epsilon^2)
\end{aligned}
\end{equation}
$$. 

What we really care is the leading order term in $\epsilon$, which is going to dictate the error of the kinetic energy. We will see it is the momentum $m v$ in this case.

#### Factorial and Stirling's Formula: How fast does $x!$ grow as $x \to \infty$?
 - One form of this says
 $
 \begin{equation}
 \ln(x!) = x \ln x - x + O(\ln x)
 \end{equation}
 $
 - A more transparent version of this will be
 $
 \begin{equation}
 x! = \sqrt{2 \pi x} (\frac{x}{e})^x (1 + O(\frac{1}{x}))
 \end{equation}
 $

The point here is to show how appropriate big $O$ is for demonstrating how fast $x!$ grows as $x \to \infty$.

#### Prime counting
We want to find the algebraic definition of the $\pi$ function where 

$$
\begin{equation}
\pi(x) = \text{number of prime numbers $\leq x$}
\end{equation}
$$.

One way to say this is

$$
\begin{equation}
\pi(x) = O(x)
\end{equation}
$$. But that's a vacuous statement, as not every integer is prime. It does not really reveal the density of prime numbers within $\mathbb{N}$.

The __Prime Number Theorem__ states

$$
\begin{equation}
\begin{aligned}{}
\pi(x) &= \text{number of prime numbers $\leq x$} \\
&= \frac{x}{\ln x} (1 + O(\frac{1}{\ln x}))
\end{aligned}
\end{equation}
$$
