**Optimization** is a class of problems for finding *what is the best*.

One of the fields of optimization is to find maxima and minima of a function.

## Finding Maxima and Minima

One observation is that if the derivative at a point of the function is not $0$, then the function is not at a maximum or minimum on that point.

And if a function has a maximum or minimum, and the function's derivative exists there, then the derivative at the maximum/minimum must be $0$.

This motivates the definition of $critical points$.

### Definition: Critical Point

A critical point of a function $f$ is an $x = a$ at which $$
\frac{d f}{d x}|_{x = a} = 0 ~ \text{or} ~ \text{undefined}
$$. A minimum is a critical point. A maximum is a critical point.

There are also critical points that are neither maxima or minima, such as an inflection.

The collection of critical points is strictly larger than the collection of extrema.

#### Classification of Critical Points

At critical point $x = a$ you have a...
- Minimum if $\frac{d^2 f}{d x^2}|_a > 0$
- Maximum if $\frac{d^2 f}{d x^2}|_a < 0$
- Fail otherwise -- you don't know if it is a maximum, minimum, or neither

Why? Think Taylor Expansion.

If we Taylor expand about the critical point $$
\begin{equation}
\begin{aligned}
f(x) &= f(a) + \frac{d f}{d x}|_a (x - a) + \frac{1}{2!} \frac{d^2 f}{d x^2}|_a (x - a)^2 + \frac{1}{3!} \frac{d^3 f}{d x^3}|_a (x - a)^3 + ... \\
&= f(a) + \frac{d f}{d x}|_a (x - a) + \frac{1}{2!} \frac{d^2 f}{d x^2}|_a (x - a)^2 + O((x - a)^3) \\
&= f(a) + \frac{1}{2!} \frac{d^2 f}{d x^2}|_a (x - a)^2 + O((x - a)^3) & \text{as $\frac{d f}{d x}|_a = 0$ at critical point $a$}
\end{aligned}
\end{equation}
$$.

So that near the critical point $a$, $f(x)$ can be approximated as a parabola. If $\frac{d^2 f}{d x^2}_a > 0$, the parabola has a minimum; if $\frac{d^2 f}{d x^2}_a < 0$, that parabola has a maximum.

```python
## Example of approximating to 2nd order derivative for sin(x)

from sympy import symbols, sin, pi, diff, plot, init_printing

init_printing()

x = symbols('x')

sin_x = sin(x)

a_maximum = pi / 2
a_minimum = 3 * pi / 2
sin_2nd_taylor_curve_maximum = sin(a_maximum) +  1 / 2 * diff(sin_x, x, 2).subs({x: a_maximum}) * (x - a_maximum) ** 2

sin_curve = plot(sin_x, xlim=(-7, 7), ylim=(-2, 2))
sin_2nd_taylor_curve_maximum_curve = plot(sin_2nd_taylor_curve_maximum, xlim=(-7, 7), ylim=(-2, 2))

sin_curve.append(sin_2nd_taylor_curve_maximum_curve[0])
sin_curve.show()
```

```python
a_minimum = 3 * pi / 2

sin_2nd_taylor_curve_minimum = sin(a_minimum) +  1 / 2 * diff(sin_x, x, 2).subs({x: a_minimum}) * (x - a_minimum) ** 2

sin_curve = plot(sin_x, xlim=(-7, 7), ylim=(-2, 2))
sin_2nd_taylor_curve_minimum_curve = plot(sin_2nd_taylor_curve_minimum, xlim=(-7, 7), ylim=(-2, 2))

sin_curve.append(sin_2nd_taylor_curve_minimum_curve[0])
sin_curve.show()
```
