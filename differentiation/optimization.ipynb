**Optimization** is a class of problems for finding *what is the best*.

One of the fields of optimization is to find maxima and minima of a function.

## Finding Maxima and Minima

One observation is that if the derivative at a point of the function is not $0$, then the function is not at a maximum or minimum on that point.

And if a function has a maximum or minimum, and the function's derivative exists there, then the derivative at the maximum/minimum must be $0$.

This motivates the definition of $critical points$.

### Definition: Critical Point

A critical point of a function $f$ is an $x = a$ at which $$
\frac{d f}{d x}|_{x = a} = 0 ~ \text{or} ~ \text{undefined}
$$. A minimum is a critical point. A maximum is a critical point.

There are also critical points that are neither maxima or minima, such as an inflection.

The collection of critical points is strictly larger than the collection of extrema.

#### Classification of Critical Points

At critical point $x = a$ you have a...
- Minimum if $\frac{d^2 f}{d x^2}|_a > 0$
- Maximum if $\frac{d^2 f}{d x^2}|_a < 0$
- Fail otherwise -- you don't know if it is a maximum, minimum, or neither

Why? Think Taylor Expansion.

If we Taylor expand about the critical point $$
\begin{equation}
\begin{aligned}
f(x) &= f(a) + \frac{d f}{d x}|_a (x - a) + \frac{1}{2!} \frac{d^2 f}{d x^2}|_a (x - a)^2 + \frac{1}{3!} \frac{d^3 f}{d x^3}|_a (x - a)^3 + ... \\
&= f(a) + \frac{d f}{d x}|_a (x - a) + \frac{1}{2!} \frac{d^2 f}{d x^2}|_a (x - a)^2 + O((x - a)^3) \\
&= f(a) + \frac{1}{2!} \frac{d^2 f}{d x^2}|_a (x - a)^2 + O((x - a)^3) & \text{as $\frac{d f}{d x}|_a = 0$ at critical point $a$}
\end{aligned}
\end{equation}
$$.

So that near the critical point $a$, $f(x)$ can be approximated as a parabola. If $\frac{d^2 f}{d x^2}_a > 0$, the parabola has a minimum; if $\frac{d^2 f}{d x^2}_a < 0$, that parabola has a maximum.

```python
## Example of approximating to 2nd order derivative for sin(x)

from sympy import symbols, sin, pi, diff, plot, init_printing

init_printing()

x = symbols('x')

sin_x = sin(x)

a_maximum = pi / 2
a_minimum = 3 * pi / 2
sin_2nd_taylor_curve_maximum = sin(a_maximum) +  1 / 2 * diff(sin_x, x, 2).subs({x: a_maximum}) * (x - a_maximum) ** 2

sin_curve = plot(sin_x, xlim=(-7, 7), ylim=(-2, 2))
sin_2nd_taylor_curve_maximum_curve = plot(sin_2nd_taylor_curve_maximum, xlim=(-7, 7), ylim=(-2, 2))

sin_curve.append(sin_2nd_taylor_curve_maximum_curve[0])
sin_curve.show()
```

```python
a_minimum = 3 * pi / 2

sin_2nd_taylor_curve_minimum = sin(a_minimum) +  1 / 2 * diff(sin_x, x, 2).subs({x: a_minimum}) * (x - a_minimum) ** 2

sin_curve = plot(sin_x, xlim=(-7, 7), ylim=(-2, 2))
sin_2nd_taylor_curve_minimum_curve = plot(sin_2nd_taylor_curve_minimum, xlim=(-7, 7), ylim=(-2, 2))

sin_curve.append(sin_2nd_taylor_curve_minimum_curve[0])
sin_curve.show()
```

### Example

Taylor Series expansion of $\sin^2(x) \ln(\cos(x))$ around critical point $x = 0$:

$$
\begin{equation}
\begin{aligned}
\sin^2(x) \ln(\cos(x)) &= (x + O(x^3))^2 \ln(1 - \frac{1}{2} x^2 + O(x^4)) \\
&= (x + O(x^3))^2 (- \frac{1}{2} x^2 + O(x^4) + \frac{1}{2} (O(x^2))^2) \\
&= (x + O(x^3))(x + O(x^3))(- \frac{1}{2} x^2 + O(x^4)) \\
&= - \frac{1}{2} x^4 + O(x^6)
\end{aligned}
\end{equation}
$$. So the $2^{nd}$ derivative is $0$, which won't tell us anything about the class of the critical point. But the $4^{th}$ derivative is negative. Hence $x = 0$ is a maximum.

### More example

Consider a square of length $L$, out of which we cut the four corners, we remove four squares, equal size, and then fold the internal edges to get an open top box. The goal is to maximize the volume of the box.

![](open_box.png)

$$
\begin{equation}
\begin{aligned}
v &= x (L - 2 x)^2 \\
&= 4 x^3 - 4 L x^2 + L^2 x
\end{aligned}
\end{equation}
$$. Hence

$$
\begin{equation}
\begin{aligned}
\frac{d V}{d x} &= 12 x^2 - 8 L x + L^2 \\
&= (6 x - L)(2 x - L)
\end{aligned}
\end{equation}
$$. From which we can know the two critical points are $$
\begin{array}{l}
x_1 &= \frac{L}{2} \\
x_2 &= \frac{L}{6}
\end{array}
$$. The maximum will be the critical point whose the $2^{nd}$ derivative will be negative. We have $$
\begin{array}{l}
\frac{d^2 V}{d x^2}|_{x_1} &= 24 x_1 - 8 L \\
&= 24 \frac{L}{2} - 8 L \\
&= 4 L \\
\frac{d^2 V}{d x^2}|_{x_2} &= 24 x_2 - 8 L \\
&= 24 \frac{L}{6} - 8 L \\
&= - 4 L
\end{array}
$$. Hence $x_2 = \frac{L}{6}$ is the maximum, and $$
\begin{equation}
\begin{aligned}
V_{max} &= 4 x_2^3 - 4 L x_2^2 + L^2 x_2 \\
&= x_2 (L - 2 x_2)^2 \\
&= \frac{L}{6} (\frac{2 L}{3})^2 \\
&= \frac{L}{6} \frac{4 L^2}{9} \\
&= \frac{2 L^3}{27}
\end{aligned}
\end{equation}
$$.

### Another example

Consider $$
f(x) = \frac{1}{3} x^3 - 3 x^2 + 8 x - 2
$$. We want to extremize $f$ for $x \geq 0$.

We know that $$
\begin{equation}
\begin{aligned}
f^\prime(x) &= x^2 - 6 x + 8 \\
&= (x - 2)(x - 4)
\end{aligned}
\end{equation}
$$. Hence the critical points are $x_1 = 2$ and $x_2 = 4$.

To classify the critical points, let's compute the $2^{nd}$ derivatives: $$
\begin{array}{l}
\frac{d^2 f}{d x^2}|_{x_1} &= 2 \cdot 2 - 6 \\
&= -2 \\
\frac{d^2 f}{d x^2}|_{x_2} &= 2 \cdot 4 - 6 \\
&= 2
\end{array}
$$. Hence at $x = x_1 = 2$ is the maximum of $f(x_1) = \frac{14}{3}$, and at $x = x_2 = 4$ is the minimum of $f(x_2) = \frac{10}{3}$.

However, $f(x_1) = \frac{14}{3}$ and $f(x_2) = \frac{10}{3}$ are **local** extrema!

### Global extrema

When asked to find a global max/min, we must check the endpoints of the domain.

For the above example, for $x = 0$, $f(0) = -2$, which is the global minimum.

And the domain is unbounded at the right-hand-side, and we can tweak the language and say $x = \infty$ is the right endpoint of the domain. And $$
\begin{equation}
\begin{aligned}
\lim_{x \to \infty} f(x) &= \lim_{x \to \infty} (x^3 - 3 x^2 + 8 x - 2) \\
&= \infty
\end{aligned}
\end{equation}
$$. Hence there is no global maximum.

Always check the endpoints.
