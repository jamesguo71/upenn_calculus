What in deed is $d x$ or $d u$?

One perspective is they are infinitesimals. Really, really small bits of $x$ or $u$.

Think $\int d u = u$.

That's an intuitive interpretation. But what do they really mean?

We can also interpret $d u$ as the rate of change. Think the chain rule.

$$
d u = \frac{d u}{d x} d x
$$

...where
- $d u$ on the left is the rate of change of $u$
- $\frac{d u}{d x}$ on the right is the rate of change of $u$ in respect to $x$
- $d x$ on the right is the rate of change of $x$

This is not the full truth, though. The full truth is they are differential forms. But we care not going to cover them in this single-variable calculus course. Sad!

## $d$ as an operator

Think $d$ is an operator...

...which means put in a function $f$ into $d$, we get out $d f$, which tells you something about the rate of change of the output respect to the input for $f$.

As $d$ is an operator, we can also put an equation as input and get an equation as output.

Think
$$
\begin{array}{l}
f &= g \\
d f &= d g
\end{array}
$$. This is often called implicit differentiation.

This is extremely useful.

### Example: Tangent line on a point of a circle

For example, if we want to know the slope of the tangent line of a circle with radius $f$.

First, we can represent the circle as $x^2 + y^2 = r^2$.

Then we can differentiate the entire equation as $d(x^2 + y^2 = r^2)$.

Hence we have $2 x d x + 2 y d y = 0$, as we treat $r$ as a constant.

Then we can obtain the slope of the tangent line as $\frac{d y}{d x} = - \frac{x}{y}$.

### Substitution (Economics)

In economics, the *marginal rate of substitution* of $x$ for $y$ is the rate at which a consumer is willing to exchange good $X$ for good $Y$. (Assuming fixed utility.)

What does it mean?

$$
MRS = - \frac{d y}{d x}
$$ along $$
u(x, y) = C
$$.

For example, let's say $u(x, y) = y^2 (x - 3)$.

Then $d u = d(y^2 (x - 3))$.

And since $u$ is a constant, then $d u = 0$.

Hence $$
\begin{equation}
\begin{aligned}
d(y^2 (x - 3)) &= 0 \\
2 y d y (x - 3) + y^2 d x &= 0
\end{aligned}
\end{equation}
$$. And now, $$
\begin{equation}
\begin{aligned}
MRS &= - \frac{d y}{d x} \\
&= \frac{y}{2 (x - 3)}
\end{aligned}
\end{equation}
$$

### Related rates

Consider a smooth stream of water from a faucet.

![](faucet.png)

How does the cross section shape change?

We assume the water stream is steady, hence the flow rate is a constant. And we assume the cross-section is circular. Let $v$ be the velocity, and $r$ be the radius of the cross-section, given the product of the area and the velocity is a constant, we have

$$
\begin{array}{l}
\pi r^2 v &= C \\
d(\pi r^2 v &= C) \\
d(r^2 v &= C) \\
2 r v d r + r^2 d v &= 0 \\
d r &= - \frac{r}{2 v} d v
\end{array}
$$

And to find out the time rate of change of $r$, we have $$
\begin{equation}
\begin{aligned}
\frac{d r}{d t} &= - \frac{r}{2 v} \frac{d v}{d t} \\
&= - \frac{r g}{2 v} &\text{as the water flow is influenced by gravity} \\
&= - \frac{r g}{2 (v_0 + g t)}
\end{aligned}
\end{equation}
$$

### Relative Rates

The normalization of differential: $\frac{d u}{u}$.

This gives the relative rate of change.

It is unit-less.

Also note $\frac{d u}{u} = d (\ln u)$.

#### Example: OHM's Law

If the voltage across a variable resistor is fixed, how are the relative rates of resistance and current fluctuations related?

OHM's Law: $$
V = I R
$$. If we differentiate the equation we have $$
d V = R d I + I d R
$$. Now we can compute the relative rates for $d V$ $$
\frac{d V}{V} = \frac{R d I + I d R}{V}
$$. SInce the voltage across a variable resistor is fixed, $d V = 0$. And since $V = I R$ $$
\begin{array}{l}
\frac{R d I + I d R}{V} = 0 \\
\frac{R d I + I d R}{I R} = 0 \\
\frac{d I}{I} + \frac{d R}{R} = 0 \\
\frac{d I}{I} = - \frac{d R}{R}
\end{array}
$$. Hence the relative rate of change of current and the relative rate of change of resistance are inversely related.

### Examples: Geometry

#### Ball

How does the relative change in volume compare to that of surface area?

In the case of a solid ball with radius $r$, the volume is $$
V = \frac{4}{3} \pi r^3
$$. When we differentiate the equation, we have $$
d V = 4 \pi r^2 d r
$$. The surface area of the ball is $$
A = 4 \pi r^2
$$. And $$
d A = 8 \pi r d r
$$. To compare of relative changes for $V$ and $A$, we compute $$
\begin{equation}
\begin{aligned}
\frac{\frac{d V}{V}}{\frac{d A}{A}} &= \frac{\frac{4 \pi r^2 d r}{\frac{4}{3} \pi r^3}}{\frac{8 \pi r d r}{4 \pi r^2}} \\
&= \frac{3 \frac{d r}{r}}{2 \frac{d r}{r}} \\
&= \frac{3}{2}
\end{aligned}
\end{equation}
$$.

#### Cube

Let there be a cube with side length $s$.

$$
\begin{array}{l}
V &= s^3 \\
d V &= 3 s^2 d s \\
A &= 6 s^2 \\
d A &= 12 s d s \\
\frac{\frac{d V}{V}}{\frac{d A}{A}} &= \frac{\frac{3 s^2 d s}{s^3}}{\frac{12 s d s}{6 s^2}} \\
&= \frac{3 \frac{d s}{s}}{2 \frac{d s}{s}} \\
&= \frac{3}{2}
\end{array}
$$.

So the relative rates of change for volume and area for both ball and square are $\frac{3}{2}$. Both what does that mean?

### Elasticity of demand

Let $P$ be the price of a good, and $Q$ is quantity of demand of the good at that price.

Then the elasticity is defined as **the rate of change of relative demand fluctuation with respect to relative price fluctuation**. $$
\begin{equation}
\begin{aligned}
E &= - \frac{\frac{d Q}{Q}}{\frac{d P}{P}} &\text{$P$ and $Q$ are inversely related} \\
\end{aligned}
\end{equation}
$$

If $0 < E < 1$, we say the demand is inelastic.

If $E > 1$, the demand is elastic.

For example, if wines are on sale, people will buy more of them. This does not work for milk.

If $E = 1$, the demand is called perfectly elastic.

#### Maximize relative revenue with respect to change of price

Let's denote the revenue as $R$, then $$
\begin{array}{l}
R &= P Q \\
d (R &= P Q) \\
d R &= d (P Q) \\
d R &= Q d P + P d Q \\
\frac{d R}{R} &= \frac{Q d P + P d Q}{R} \\
&= \frac{Q d P + P d Q}{P Q} \\
&= \frac{d P}{P} + \frac{d Q}{Q} \\
\frac{\frac{d R}{R}}{\frac{d P}{P}} &= \frac{\frac{d P}{P}}{\frac{d P}{P}} + \frac{\frac{d Q}{Q}}{\frac{d P}{P}} \\
&= 1 - E
\end{array}
$$

When $E = - \frac{\frac{d Q}{Q}}{\frac{d P}{P}} = 1$, the relative change of price does not affect the relative change of revenue.
