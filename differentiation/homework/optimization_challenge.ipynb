## 1. 

In this problem we will study a very basic traffic model. Consider a stretch of highway in which cars are traveling at an average speed $v$. The *traffic density* $u$ is the total amount of cars on our stretch of road divided by its length. These two quantities are related: the less cars on the road, the faster drivers are able to go. On the other hand, if traffic becomes heavy, drivers will naturally decrease their speed. The so-called *parabolic model* assumes that this relationship is dictated by the equation $$
u = u_{max} (1 - \frac{v}{v_{max}})
$$ where $v_{max}$ represents the capacity of the road, and $v_{max}$ the speed limit on it.

The amount of cars passing through our road is called the *traffic flux* or *throughput*, and is given by the product of the traffic density and the average speed: $$
F = u v
$$ Using the parabolic model, find out at what average speed $v_*$, the flux through our road is maximized.

### Solution

$$
\begin{equation}
\begin{aligned}
F &= u v \\
&= u_{max} (1 - \frac{v}{v_{max}}) v \\
&= u_{max} v - u_{max} \frac{v^2}{v_{max}}
\end{aligned}
\end{equation}
$$. Hence $$
\begin{equation}
\begin{aligned}
\frac{d F}{d v} &= u_{max} - 2 v \frac{u_{max}}{v_{max}}
\end{aligned}
\end{equation}
$$. To compute the root of $\frac{d F}{d v}$: $$
\begin{array}{lcl}
\frac{d F}{d v} &= 0 \\
u_{max} - 2 v \frac{u_{max}}{v_{max}} &= 0 \\
2 v \frac{u_{max}}{v_{max}} &= u_{max} \\
v &= \frac{v_{max}}{2}
\end{array}
$$. Hence $v = \frac{v_{max}}{2}$ is a critical point of $F$.

And computing the $2^{nd}$ order derivative of $F$: $$
\frac{d^2 F}{d v^2} = - 2 \frac{u_{max}}{v_{max}} < 0
$$... which is a negative constant. Since there is only one critical point, then at $v = \frac{v_{max}}{2}$ there is the global maximum of $F$. Hence $$
V_* = \frac{v_{max}}{2}
$$

## 2. 

A manufacturing company wants to know how many workers it should hire. If it employs too many people, the machines in the factory will be overutilized and the workers will have to wait until they are free, thus reducing the number of units each one will produce in a day's work. On the other hand, too few workers would leave the machines idle for long periods of time.

A rough model for the relationship between the number $n$ of workers and their productivity $p$ is given by the equation $$
p = p_{max} (1 - \frac{n}{n_{max}})
$$

where $p_{max} = 10$ is the maximum number of units a worker can produce in a day and $n_{max} = 100$ is the maximum number of workers the factory can accommodate.

The amount of units $U$ manufactured in the whole factory in one day is equal to the product of the number of workers and the number of units each one produces: $$
U = n p
$$ How many workers should the company hire in order to maximize its production?

### Solution

To find the critical points of $U$, we need to compute the $1^{st}$ derivative of $U$: $$
\begin{equation}
\begin{aligned}
\frac{d U}{d n} &= \frac{d}{d n} (n p) \\
&= \frac{d}{d n} (n p_{max} (1 - \frac{n}{n_{max}})) \\
&= \frac{d}{d n} (p_{max} n - \frac{p_{max}}{n_{max}} n^2) \\
&= p_{max} - \frac{2 p_{max}}{n_{max}} n
\end{aligned}
\end{equation}
$$. To find the critical point of $U$, we also need to find the root of $\frac{d U}{d n}$: $$
\begin{array}{l}
\frac{d U}{d n} &= 0 \\
p_{max} - \frac{2 p_{max}}{n_{max}} n &= 0 \\
\frac{2 p_{max}}{n_{max}} n &= p_{max} \\
n &= \frac{n_{max}}{2}
\end{array}
$$. And to classify the only critical point we need to compute the second order derivative: $$
\frac{d^2 U}{d n^2} = - \frac{2 p_{max}}{n_{max}} < 0
$$. Which is a negative constant. Hence $U$ at $n = \frac{n_{max}}{2}$ we have the global maximum.

By plugging in and $n_{max} = 100$, we have the global maximum as $$
n = 50
$$.

## 3.
A technology company has just invented a new gadget. In order to maximize the profit derived from its sale, the company must make a critical decision: at what price should it be sold? A market study suggests that the number $N$ of units sold would approximately follow the equation $$
N = N_{max} e^{-\frac{P}{\lambda}}
$$

where $P$ is the sale price, $N_{max} = 10, 000, 000$ is the number of units that would saturate the market, and $\lambda = \$50$.

If it costs $\$250$ to manufacture one of these gadgets, at what price $P_*$ would be profit of the company be maximized?

### Solutions

Let's define the profit function $U$

$$
\begin{equation}
\begin{aligned}
U &= N (P - 250) \\
&= N_{max} e^{-\frac{P}{\lambda}} (P - 250) \\
&= P N_{max} e^{-\frac{P}{\lambda}} - 250 N_{max} e^{-\frac{P}{\lambda}}
\end{aligned}
\end{equation}
$$

To find the global maximum of $U$ for $P$ in $[0, \infty]$.

First we need to compute the derivative of $U$:

$$
\begin{equation}
\begin{aligned}
\frac{d}{d P} P N &= (N_{max} e^{-\frac{P}{\lambda}} - N_{max} \frac{P}{\lambda} e^{-\frac{P}{\lambda}}) - 250 N_{max} e^{-\frac{P}{\lambda}} (- \frac{1}{\lambda}) \\
&= N_{max} e^{-\frac{P}{\lambda}} (1 - \frac{P}{\lambda} + \frac{250}{\lambda})
\end{aligned}
\end{equation}
$$ And to find the root of the derivative:

$$
\begin{array}{l}
\frac{d}{d P} U &= 0\\
(1 - \frac{P}{\lambda} + \frac{250}{\lambda}) &= 0 \\
\frac{P - 250}{\lambda} &= 1 \\
P &= \lambda + 250 \\
&= 300
\end{array}
$$.

And computing the $2^{nd}$ derivative of $U$ $$
\begin{equation}
\begin{aligned}
\frac{d^2}{d P^2} P N &= - \frac{N_{max}}{\lambda} e^{-\frac{P}{\lambda}} - (N_{max} \frac{1}{\lambda} e^{-\frac{P}{\lambda}} - N_{max} \frac{P}{\lambda^2} e^{-\frac{P}{\lambda}})) - 250 N_{max} e^{-\frac{P}{\lambda}} (- \frac{1}{\lambda})^2 \\
&= - \frac{N_{max}}{\lambda} e^{-\frac{P}{\lambda}} (1 - \frac{1}{\lambda} + \frac{P}{\lambda^2} - \frac{250}{\lambda^2}) \\
&= - \frac{N_{max}}{\lambda} e^{-\frac{P}{\lambda}} (\frac{\lambda^2 - \lambda + P - 250}{\lambda^2})
\end{aligned}
\end{equation}
$$

By plugging $x = 300$, we have $\frac{d^2}{d P^2} U|_{p = 300} < 0$. Hence $P_* = \$ 300$
