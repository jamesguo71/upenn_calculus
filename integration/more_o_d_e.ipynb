We will consider more families of differential equations in this lesson.

## More families of ODEs
### Autonomous
$\frac{\operatorname{d} x}{\operatorname{d} t} = f(x)$, as the derivative only depends on $x$.
### Non-autonomous
$\frac{\operatorname{d} x}{\operatorname{d} t} = f(x, t)$, which are usually very difficult to solve. Some of them however, are solvable.

### Separable

$\frac{\operatorname{d} x}{\operatorname{d} t} = f(x) g(t)$. Then following the pattern of linear differentiation, we can separate the ODE as $$
\begin{array}{l}
\frac{\operatorname{d} x}{f(x)} &= g(t) \operatorname{d} t \\
\int \frac{1}{f(x)} \operatorname{d} x &= \int g(t) \operatorname{d} t \\
\end{array}
$$

### Example: Drag force

Assume a drag force proportional to velocity. What is the velocity of the body at time $t$?

Let $m$ be the mass, and $g$ be the gravitational constant, then the gravitational force the body is subject to is $m g$.

Meanwhile, the drag force is $- \kappa v$, where $\kappa$ is a constant. Then we have $$
m \frac{\operatorname{d} v}{\operatorname{d} t} = m g - \kappa v
$$. This ODE is *autonomous* and *separable*. By doing a little factoring we have $$
\begin{array}{l}
m \frac{\operatorname{d} v}{\operatorname{d} t} &= m g - \kappa v \\
m \frac{\operatorname{d} v}{\operatorname{d} t} &= \kappa (\frac{m g}{\kappa} - v) \\
\frac{\operatorname{d} v}{\frac{m g}{\kappa} - v} &= \frac{\kappa}{m} \operatorname{d} t \\
\frac{\operatorname{d} v}{v - \frac{m g}{\kappa}} &= - \frac{\kappa}{m} \operatorname{d} t \\
\int \frac{1}{v - \frac{m g}{\kappa}} \operatorname{d} v &= \int - \frac{\kappa}{m} \operatorname{d} t \\
\ln(v - \frac{m g}{\kappa}) &= - \frac{\kappa}{m} t + C \\
v - \frac{m g}{\kappa} &= e^{- \frac{\kappa}{m} t + C} \\
&= C e^{- \frac{\kappa}{m} t} \\
v &= C e^{- \frac{\kappa}{m} t} + \frac{m g}{\kappa}
\end{array}
$$. What if $t$ goes to infinity? $$
\begin{equation}
\begin{aligned}
\lim_{t \to \infty} v(t) &= \lim_{t \to \infty} (C e^{- \frac{\kappa}{m} t} + \frac{m g}{\kappa}) \\
&= 0 + \frac{m g}{\kappa} \\
&= \frac{m g}{\kappa}
\end{aligned}
\end{equation}
$$. This is called the *terminal velocity*.

### Example: Torricelli's law

This model is for the fluid leaking from the tank.

![](torricellis_law.png)

$$
\begin{equation}
\begin{aligned}
\frac{\operatorname{d} h}{\operatorname{d} t} &= - \frac{\kappa}{A} \sqrt{h} \\
&= - \frac{\kappa}{\pi h^2} \sqrt{h} & \text{As the tank is a cone which area $A$ is proportional to $h$} \\
&= - \frac{\kappa}{\pi} h^{-\frac{3}{2}} \\
h^{\frac{3}{2}} \operatorname{d} h &=  - \frac{\kappa}{\pi} \operatorname{d} t \\
\int h^{\frac{3}{2}} \operatorname{d} h &= \int - \frac{\kappa}{\pi} \operatorname{d} t \\
\frac{2}{5} h^{\frac{5}{2}} &= - \frac{\kappa}{\pi} t + C \\
h &= (- \frac{5 \kappa}{2 \pi} t + C)^{\frac{2}{5}}
\end{aligned}
\end{equation}
$$. What that implies is $h$ approaches to $0$ very rapidly: $O((t_{end} - t)^{\frac{2}{5}})$.

### Linear *non-autonomous* ODE'S

$$
\frac{\operatorname{d} x}{\operatorname{d} t} = A(t) x + B(t)
$$. We mean *linear* as if we suppress $t$ dependent, we have $$
\frac{\operatorname{d} x}{\operatorname{d} t} = A x + B
$$

For example $$
\frac{\operatorname{d} x}{\operatorname{d} t} = t x + \frac{1}{2} t^2
$$. How do we solve such kind of ODE's? The method we are going to introduce is called **integrating factor**, which we denote by $\operatorname{I}$. $$
\begin{array}{l}
\operatorname{I}(\frac{\operatorname{d} x}{\operatorname{d} t} &= A x + B) \\
\operatorname{I}(\frac{\operatorname{d} x}{\operatorname{d} t} - A x &= B) \\
\operatorname{I} \frac{\operatorname{d} x}{\operatorname{d} t} - A \operatorname{I} x &= \operatorname{I} B
\end{array}
$$. Let's pause for a moment. If we can transform the above equation into $$
\operatorname{I} \frac{\operatorname{d} x}{\operatorname{d} t} + \frac{\operatorname{d} \operatorname{I}}{\operatorname{d} t} x = \operatorname{I} B
$$, then from the product rule, we will have $$
\operatorname{d} (\operatorname{I} x) = \operatorname{I} B
$$. So what needs to be true to have the above equation holds? We will need $$
\begin{array}{l}
- A \operatorname{I} x &= \frac{\operatorname{d} \operatorname{I}}{\operatorname{d} t} x \\
\frac{\operatorname{d} \operatorname{I}}{\operatorname{d} t} &= - A \operatorname{I}
\end{array}
$$. And this equation is *separable*, which means $$
\begin{array}{l}
\frac{\operatorname{d} \operatorname{I}}{\operatorname{d} t} &= - A \operatorname{I} \\
\frac{\operatorname{d} \operatorname{I}}{\operatorname{I}} &= - A \operatorname{d} t \\
\int (\frac{\operatorname{d} \operatorname{I}}{\operatorname{I}} &= - A \operatorname{d} t) \\
\int \frac{1}{\operatorname{I}} \operatorname{d} \operatorname{I} &= - \int A \operatorname{d} t \\
\ln(\operatorname{I}) &= - \int A\\
\operatorname{I} &= e^{- \int A }
\end{array}
$$. So we choose $I$ to be $$
\operatorname{I} = e^{- \int A}
$$, then we will have $$
\begin{array}{l}
\operatorname{d} (\operatorname{I} x) &= \operatorname{I} B \\
\int (\operatorname{d} (\operatorname{I} x) &= \operatorname{I} B) \\
\int \operatorname{d} (\operatorname{I} x) &= \int \operatorname{I} B \\
\operatorname{x} &= \frac{1}{\operatorname{I}} \int \operatorname{I} B \\
&= e^{\int A} (\int B e^{- \int A}) \\
&= e^{\int A(t) \operatorname{d} t} (\int B(t) e^{- \int A(t)} \operatorname{d} t) \\
\end{array}
$$

#### Example: Mixing

A $1000$ gallon tank is $90\%$ full.

An additive is injected into the tank at rate $10$ gallon/minute.

The mixture is well-stripped and drained at rate of $5$ gallon/minute.

What is the concentration of the additive when the tank is full?

Let
- $V(t)$ be the volume at time $t$.
- $Q(t)$ be the quantity of the additive at time $t$.
- $C(t)$ be the concentration at time $t$, that $C(t) = \frac{Q(t)}{V(t)}$.

Then we have $$
V(t) = 900 + (10 - 5) t
$$, and we then know the tank will be full in 20 minutes.

We also know that $$
\begin{array}{l}
\frac{\operatorname{d}}{\operatorname{d} t} Q = 10 - 5 C \\
&= 10 - 5 \frac{Q}{V} \\
&= 10 - \frac{Q}{180 + t} \\
\frac{\operatorname{d}}{\operatorname{d} t} Q + \frac{Q}{180 + t} &= 10
\end{array}
$$. That means the integrating factor $\operatorname{I}$ is $$
\begin{array}{l}
\operatorname{I} &= e^{\int \frac{1}{180 + t} \operatorname{d} t} \\
&= e^{\ln(180 + t)} \\
&= 180 + t
\end{array}
$$. Then we can multiply the original ODE by the integrating factor $\operatorname{I}$: $$
\begin{array}{l}
\frac{\operatorname{d}}{\operatorname{d} t} Q &= 10 - \frac{Q}{180 + t} \\
\operatorname{I} \frac{\operatorname{d}}{\operatorname{d} t} Q + \operatorname{I} \frac{Q}{180 + t} &= \operatorname{I} \cdot 10 \\
(180 + t) \frac{\operatorname{d}}{\operatorname{d} t} Q + (180 + t) \frac{Q}{180 + t} &= (180 + t) \cdot 10 \\
(180 + t) \frac{\operatorname{d}}{\operatorname{d} t} Q + Q &= (180 + t) \cdot 10 \\
\operatorname{d} ((180 + t) Q) &= 10 (180 + t) \\
\int \operatorname{d} ((180 + t) Q) &= \int 10 (180 + t) \operatorname{d} t \\
(180 + t) Q &= 1800 t + 5 t^2 + C \\
(180 + t) Q &= 1800 t + 5 t^2 + 0 &\text{as the initial quantity of the additive is $0$} \\
Q &= \frac{1800 t + 5 t^2}{180 + t} \\
Q(20) &= \frac{1800 t + 5 t^2}{180 + t}|_{t = 20} \\
&= 190 \\
C(20) &= \frac{V(20)}{Q(20)} \\
&= \frac{190}{1000} \\
&= 19 \%
\end{array}
$$
